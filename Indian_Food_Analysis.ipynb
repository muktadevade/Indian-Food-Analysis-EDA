The dataset consists of about **255 Indian dishes** and **9** columns associated with each of them.
The 9 columns are as follows
1. **name**        - name of the dish
2. **ingredients** - main ingredients used
3. **diet**        - type of diet (vegetarian , non-vegetarian )
4. **Prep_time**   - preparation time
5. **cook_time**   - cooking time
6. **flavor_profile** -flavor profile includes whether the dish is spicy,sweet,bitter etc
7. **course** - course of meal - starter, main course , dessert etc
8.**state** - state where the dish is famous or is originated
9. region - region where the state belongs

```python
pie_data = data.diet.value_counts().reset_index()
```

```python
pie_data.columns = ['diet','count']
fig = px.pie(pie_data, values = 'count', names='diet',title='prportion of vegetarian and non vegetarian dishes',
color_discrete_sequence=['green','red'])
fig.show()
```

```python
sweet_data = data[data['flavor_profile']=='sweet']
final_sweet_data = sweet_data[sweet_data['course']!='dessert']
```

```python
final_sweet_data
```

```python
flav_data = data.flavor_profile.value_counts().reset_index()
flav_data.columns = ['flavor_profile', 'prep_time']
fig = px.bar(flav_data,x='flavor_profile',y='prep_time',title='variety of item according to the flavour',
color_discrete_sequence=['green'])
fig.show()
```
